```python
import json
import networkx as nx
from pathlib import Path

WORKING_DIR = "./nano_graphrag_cache_deepseek_TEST"
GRAPHML = Path(WORKING_DIR) / "graph_chunk_entity_relation.graphml"
OUT = Path("graph_json.js")  # se guardará junto al HTML

G = nx.read_graphml(GRAPHML)

nodes = []
for nid, data in G.nodes(data=True):
    nodes.append({
        "id": data.get("entity_name") or nid,              # etiqueta visible
        "entity_type": data.get("entity_type", "unknown"), # para colorear
        "description": data.get("description", "")         # para tooltip
    })

links = []
for u, v, data in G.edges(data=True):
    links.append({
        "source": G.nodes[u].get("entity_name") or u,
        "target": G.nodes[v].get("entity_name") or v,
        "value": int(data.get("relationship_strength", 1)),  # grosor de arista
        "description": data.get("description", "")
    })

payload = {"nodes": nodes, "links": links}
with open(OUT, "w", encoding="utf-8") as f:
    f.write("const graphJson = ")
    json.dump(payload, f, ensure_ascii=False)
    f.write(";")
print("✅ Escrito", OUT.resolve())
```
